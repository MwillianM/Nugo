```python
import numpy as np
import pandas as pd
from apyori import apriori, dump_as_json
```

```python
def f(df):
    keys, values = df.sort_values('ID').values.T
    ukeys, index = np.unique(keys,True)
    arrays = np.split(values,index[1:])
    df2 = pd.DataFrame({'ID':ukeys,'Produto':[list(a) for a in arrays]})
    return df2

def unpack(rr):
    os = rr.ordered_statistics[0]
    return ' e '.join([*os.items_base,*os.items_add])
```

```python
df = pd.read_csv('Produtos.csv', dtype={'ID':np.int32})
df.head()
```

```python
df = f(df)
df.tail(15)
```

```python
df.Produto.apply(len).value_counts()[:10]
```

```python
df = df[df['Produto'].apply(len) > 1]
df.head()
```

```python
len(df.index)
```

```python
results = list(apriori(list(df.Produto), min_support=0.0009, min_confidence=0.6))
for i in range(len(results)):
    print('Regra ', i+1, ':')
    print('  Items: ', unpack(results[i]))
    print('  Suporte: ', int(results[i].support*len(df.index)))
    print('  Confiança: %.2f' %results[i].ordered_statistics[0].confidence)
```
